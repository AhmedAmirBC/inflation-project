Inflation data Scraper
------------------------
_______________________________
Data from https://www.inflation.eu/inflation-rates

CPI-U for various countries. The inflation "rate" is stated, which is calculated as 
$$\dfrac{CPI_t - CPI_{t - 1}}{CPI_{t - 1}}$$ 
for time $t$.

```python
import matplotlib.pyplot as plt
import bs4
import requests
```

```python
# I have no idea why I can't just import bs4.BeautifulSoup as bs4!
bs = bs4.BeautifulSoup
```

### Create a list of the relevant urls

```python
def create_url(year):
    url_base = "https://www.inflation.eu/inflation-rates/"
    year = str(year)
    return url_base + f"cpi-inflation-{year}" + ".aspx"

start_year = 1973
end_year = 2018 ## modify as needed. Leave as current year - 1.
url_list = dict({})

for year in range(start_year, end_year + 1):
    url_list[year] = create_url(year)
```

```python
url_list[start_year], url_list[end_year]
```

### Pull the percentage rates from pages

```python
data = dict({})

def pull_data(countries, y):
    countries_this_year = [pull_name(country[1]) for country in countries]
    rates_this_year = [pull_rate(country[-2]) for country in countries]
    return countries_this_year, rates_this_year

def pull_name(td):
    return td.text[15:-5]
    
def pull_rate(td):
    rate = td.text[:-3]
    rate_parts = rate.split(".")
    if len(rate_parts) == 2:
        return float(rate)
    else:
        rate = float(str(rate_parts[0]) + str(rate_parts[1]) + "." + str(rate_parts[-1]))
        return rate

def compile_(d, c, r, y):
    number_of_countries = len(c)
    index = 0
    while index < number_of_countries:
        if c[index] in d:
            d[c[index]][y] = r[index]
        else:
            d[c[index]] = dict({y: r[index]})
        index += 1
```

```python
## Tests

source = requests.get(url_list[start_year])
sauce = source.text
source.close()
soup = bs(sauce, "html5")

countries = soup.find_all(class_ = "tabledata1") + soup.find_all(class_ = "tabledata2")
countries = [country.contents for country in countries]
c, r = pull_data(countries, start_year)
print(c, r)
```

```python
for year in range(start_year, end_year + 1):
    source = requests.get(url_list[year])
    sauce = source.text
    source.close()
    
    soup = bs(sauce, "html5")
    countries_this_year = []
    rates_this_year = []
    
    odd_countries = soup.find_all(class_ = "tabledata1")
    even_countries = soup.find_all(class_ = "tabledata2")
    countries = odd_countries + even_countries
    countries = [country.contents for country in countries]
    
    countries_this_year, rates_this_year = pull_data(countries, year)
    print(countries_this_year, rates_this_year)
    compile_(data, countries_this_year, rates_this_year, year)
    print(data)
    
print(data)
```

```python
aus = data["Austria"]
s = min(aus.keys())
e = max(aus.keys())
aus_rates = [aus[i] for i in range(s, e + 1)]
aus_years = [i for i in range(s, e + 1)]
print(aus_rates)
print(aus_years)
plt.scatter(aus_rates, aus_years)
```
