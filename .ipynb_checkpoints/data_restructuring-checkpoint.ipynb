```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open("inflation_data.pickle", "rb") as handle:
    data = pickle.load(handle)
print(data["United States"])
```

```python
def create_time_series(rates):
    """
    Creates a time series of inflation rates every year
        inputs: 
            - rates: dict with key-value pairs of "year"-"annual CPI Inflation". Expects that years recorded are consecutive.
        returns:
            - years: x-axis of a possible inflation plot
            - rates: y-axis of a possible inflation plot
    
    Possible bug: Incomplete inputs may result in some years not being a key in the dictionary. 
    """
    start = min(rates.keys())
    end = max(rates.keys())
    
    years = [i for i in range(start, end + 1)]
    dates = [rates[i] for i in years]
    return years, dates
```

```python
countries = data.keys()

# data restructured for convenient plotting
restructured_data = dict({})
for country in countries:
    restructured_data[country] = create_time_series(data[country])

print(restructured_data["United States"])
```

```python
def plot(country, name):
    """
    Unpacks data into rates and years. Plots it.
    
    Possible extensions:
        - Support for plotting multiple countries.
        ON THIS NOTE: calling plot multiple times for different countries followed by "plt.legend(loc='upper right')" will
        work beautifully. It does seem a bit of a hack though.
    """
    rates, years = country
    plt.plot(rates, years, label=name)
```

```python
plot(restructured_data["United States"])
```

### Data is ready for analysis

Below is one possible analysis. We study which time series shows the most variance, and analyse variance with respect to US CPI Inflation rates

#### Our Sample

We will study inflation particularly for the period from 1999-2018, wherein all of the countries in our dataset have inflation data. We will first select this data in particular.

```python
import numpy as np
```

```python
years_covered_by_all = {i for i in range(1973, 2019)}

for country in restructured_data.keys():
    years_covered_by_all = years_covered_by_all.intersection(restructured_data[country][0])
    
years_covered_by_all = [i for i in range(min(years_covered_by_all), max(years_covered_by_all) + 1)]
print(years_covered_by_all)
```

```python
def combine_countries(data, country_list):
    """
    Aggregates data for various countries. Selects only the years for which all data is available. 
    This makes variance analysis possible.
    
    Possible extensions:
        - Could try to plot data without having to have only common years represented.
        - Could try to extend to take care of missing years. This is unnecessary in our (mercifully) complete dataset.
    """
    years_covered_by_all = {i for i in range(1973, 2019)}
    for country in country_list:
        years_covered_by_all = years_covered_by_all.intersection(data[country][0])
    years_covered_by_all = [i for i in range(min(years_covered_by_all), max(years_covered_by_all) + 1)]
    range_years = len(years_covered_by_all)
    
    combined_data = dict({})
    for country in country_list:
        combined_data[country] = years_covered_by_all, data[country][1][-range_years:]
        assert len(combined_data[country][1]) == range_years
    return combined_data

test_combination = combine_countries(restructured_data, ["United States", "Great Britain", "France", "Turkey"])
```

```python
print(test_combination)
```

```python
for country in test_combination:
    plot(test_combination[country], country)
plt.legend(loc="upper left")
```
